```python
import numpy as np
import math
import matplotlib.pyplot as plt

class BrownianMotion:
    """
    A class to sample a BrownianMotion at time t.

    ...

    Attributes
    ----------
    drift : num
        dirft of the brownian motion
    mean : 0
        mean of the normal distribution
    sd : num
        standard deviation of Brownian Motion
    turn: int
        turn of the game

    Methods
    -------
    brown_mo(BrownianMotion):
        samples the brownian motion with drift at turn
    """
    def __init__(self, drift, sd, turn):
        self.drift = drift
        self.sd = sd
        self.turn = turn

    def brown_mo(self):
        drift = self.drift
        sd = self.sd
        turn = self.turn 
        
        time_hor = turn
        if turn < 100:
            n = turn * 100
        else:
            n = turn * 20
        
        time_step = time_hor/n
        time = time_step * np.ones(n)

        normal_sample = np.random.normal(0, 1, n) * math.sqrt(time_step)
        BM_path = drift * time + sd* np.cumsum(normal_sample)
        return BM_path
            
alumimum = BrownianMotion(2500, 1000, 10000)
alumimum_bm = alumimum.brown_mo()

plt.plot(range(0, 10000*20), alumimum_bm)
```
