```python
import numpy as np
import math
import matplotlib.pyplot as plt
'''
The following class samples brownian motion at time t
'''
class BrownianMotion:
    def __init__(self, drift, mean, sd, turn):
        self.drift = drift
        self.mean = mean
        self.sd = sd
        self.turn = turn

    def brown_mo(self):
        drift = self.drift
        mean = self.mean
        sd = self.sd
        turn = self.turn 
        
        time_hor = turn 
        if turn < 100:
            n = turn * 100
        else:
            n = turn * 20
        
        time_step = time_hor/n
        time = time_step * np.ones(n)

        normal_sample = np.random.normal(0, 1, n) * math.sqrt(time_step)
        BM_path = drift * time + sd* np.cumsum(normal_sample)
        return BM_path
            
alumimum = BrownianMotion(10000, 0, 1000, 50)
alumimum_bm = alumimum.brown_mo()

plt.plot(range(0, 50*100), alumimum_bm)
```
