```python
import numpy as np
import math
import matplotlib.pyplot as plt
'''
The following class samples brownian motion at time t
'''
class BrownianMotion:
    def __init__(self, drift, mean, sd, turn):
        self.drift = drift
        self.mean = mean
        self.sd = sd
        self.turn = turn

    def brown_mo(self):
        drift = self.drift
        mean = self.mean
        sd = self.sd
        turn = self.turn 
        
        time_hor = turn 
        if turn < 100:
            n = turn * 100
        else:
            n = turn * 20
        
        time_step = []
        for i in range(1, n + 1):
            time_step.append(i * time_hor/n)

        normal_sample = np.random.normal(mean, sd, n) * math.sqrt(time_hor/n)
        print(len(normal_sample))
        print(turn)
        BM_path = np.cumsum(normal_sample)
        return BM_path
            
alumimum = BrownianMotion(0, 0, 1, 50)
alumimum_bm = alumimum.brown_mo()

plt.plot(range(0, 50*100), alumimum_bm)
```
